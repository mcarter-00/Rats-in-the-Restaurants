```python
# Initial imports.
import pandas as pd
import numpy as np


from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

pd.set_option('display.max_columns', 999)
```

```python
file_path = "Resources/clean_new_rats_JUNE09.csv"
restaurant_list_df = pd.read_csv(file_path)
restaurant_list_df
```

```python
rats_ml_df=restaurant_list_df.loc[restaurant_list_df["FACILITY_CITY"]!="LOS ANGELES, CITY OF" ]

rats_ml_df
```

```python
#drop columns
rats2_ml_df=rats_ml_df.drop(["Unnamed: 0","ACTIVITY_DATE","ACTIVITY_DATE_MONTHYEAR","serial_number","FACILITY_ID","FACILITY_NAME","PROGRAM_NAME","PROGRAM_STATUS","FACILITY_ADDRESS","FACILITY_STATE","FACILITY_CITY","GRADE","violation_status","violation_code","violation_description","GEONAME"],1)
rats2_ml_df
```

```python
# Define the features set.
X = rats2_ml_df.copy()
X = X.drop("SCORE", axis=1)
X.head()
```

```python
# Define the target set.
y = rats2_ml_df["SCORE"].ravel()
y[:5]
```

```python
# Splitting into Train and Test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=128, random_state=78) 
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data.
predictions = rf_model.predict(X_test_scaled)
```

```python
# Calculating the confusion matrix.
cm = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1","Actual 2","Actual 3","Actual 4", "Actual 5","Actual 6","Actual 7", "Actual 8", "Actual 9","Actual 10","Actual 11","Actual 12", "Actual 13","Actual 14","Actual 15", "Actual 16", "Actual 17","Actual 18","Actual 19","Actual 20", "Actual 21","Actual 22","Actual 23", "Actual 24", "Actual 25","Actual 26","Actual 27","Actual 28", "Actual 29","Actual 30"], columns=[ "Predicted 0", "Predicted 1","Predicted 2","Predicted 3","Predicted 4", "Predicted 5","Predicted 6","Predicted 7", "Predicted 8", "Predicted 9","Predicted 10","Predicted 11","Predicted 12", "Predicted13","Predicted 14","Predicted 15", "Predicted 16", "Predicted 17","Predicted 18","Predicted 19","Predicted 20", "Predicted 21","Predicted 22","Predicted 23", "Predicted 24", "Predicted 25","Predicted 26","Predicted 27","Predicted 28", "Predicted 29","Predicted 30"])

cm_df
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
#print("Confusion Matrix")
#display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```
