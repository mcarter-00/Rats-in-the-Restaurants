# First Iteration
It appends longitude & latitude coordinates to the LA County Health Profile CSV using USZipCA.csv based on GeoNames
This was before Esther's lookup table was known, so this will be converted to use that lookup table in the next ieration.

```python
if bNew:
    dfC = pd.read_csv("data/USZipsCA.csv")
else:
    dfC = pd.read_csv("zip_codes_CA.csv")
sListDuplicates = []
dfC.head()
```

```python
dfGeo = pd.read_csv("data/Los_Angeles_County_City_and_Community_Health_Profiles_2018.csv")
dfGeo.head()
```

```python
empty_dataframe = pd.DataFrame()
#list_dataframe = pd.DataFrame([1, 2, 3])
#print(result_dataframe)

if (bNew):
    sLabelCounty = 'county_name' #'county'
    sLabelLat = 'lat' #'latitude'
    sLabelLng = 'lng' #'longitude'
else:
    sLabelCounty = 'county'
    sLabelLat = 'latitude'
    sLabelLng = 'longitude'
    
for i, row in dfGeo.iterrows():
    sGeoName = row['GEONAME']
    #dfC.loc[dfC['county'] == sGeoName].count()
    if (dfC.loc[dfC[sLabelCounty] == sGeoName].count()[sLabelCounty] > 1):
        #sListDuplicates.Add(dfC.loc[dfC['county'] == sGeoName])
        result_dataframe = empty_dataframe.append(dfC.loc[dfC[sLabelCounty] == sGeoName])
        dlat = dfC.loc[dfC[sLabelCounty] == sGeoName][sLabelLat].values[0]
        dlng = dfC.loc[dfC[sLabelCounty] == sGeoName][sLabelLng].values[0]
    elif (dfC.loc[dfC[sLabelCounty] == sGeoName].count()[sLabelCounty] > 0):    
        dlat = dfC.loc[dfC[sLabelCounty] == sGeoName][sLabelLat].values[0]
        dlng = dfC.loc[dfC[sLabelCounty] == sGeoName][sLabelLng].values[0]
    else:
        if (dfC.loc[dfC['city'] == sGeoName].count()['city'] > 1):
            #sListDuplicates.Add(dfC.loc[dfC['county'] == sGeoName])
            result_dataframe = empty_dataframe.append(dfC.loc[dfC['city'] == sGeoName])
            dlat = dfC.loc[dfC['city'] == sGeoName][sLabelLat].values[0]
            dlng = dfC.loc[dfC['city'] == sGeoName][sLabelLng].values[0]
        elif (dfC.loc[dfC['city'] == sGeoName].count()['city'] > 0):

            dlat = dfC.loc[dfC['city'] == sGeoName][sLabelLat].values[0]
            dlng = dfC.loc[dfC['city'] == sGeoName][sLabelLng].values[0]
        else:
            dlat = -1
            dlng = -1
        #dlat = -1
        #dlng = -1
    #lng = row['longitude']
    #city = citipy.nearest_city(lat, lng).city_name
    dfGeo.at[i, sLabelLat] = dlat
    dfGeo.at[i, sLabelLng] = dlng
                                                  
    #df3.set_value(i, 'City', city)
    print(str(dlat) + ', ' + str(dlng) + ': ' + sGeoName)

#dfC
```
