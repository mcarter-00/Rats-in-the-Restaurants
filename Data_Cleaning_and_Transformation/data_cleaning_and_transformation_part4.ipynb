### Data cleaning accomplished in this notebook:
* Dropping columns (a lot)
* Setting facility city = "Los Angeles, City of"


```python
import pandas as pd

# Expand output display to view all
pd.set_option('display.max_columns', 999)
```

```python
dataset = pd.read_csv("./Resources/Clean/clean_new_rats_JUNE09.csv")
rats_df = pd.DataFrame(dataset)
rats_df.head()
```

```python
revised_rats_df = rats_df.loc[rats_df['FACILITY_CITY'] != 'LOS ANGELES, CITY OF']
revised_rats_df.head()
```

```python
new_rats_df = revised_rats_df.drop(columns= ["Unnamed: 0" , "ACTIVITY_DATE", "ACTIVITY_DATE_MONTHYEAR", 
                                             "FACILITY_ID", "FACILITY_NAME", "PROGRAM_NAME", "PROGRAM_STATUS", 
                                             "FACILITY_ADDRESS", "FACILITY_CITY", "FACILITY_STATE", "GRADE", 
                                             "serial_number", "violation_status", "violation_code", 
                                             "violation_description", "GEONAME"])
new_rats_df
```

NOTE: new_rats_df now has 44 columns

```python
new_rats_df.to_csv("./Resources/Clean/clean_rats_JUNE10.csv")
```
