### Data cleaning accomplished in this notebook:
* Replacing values that have ** with 0
* Dropping "Unnamed: 0.1"

```python
# Import dependencies
import pandas as pd
```

```python
# Load dataset
dataset = pd.read_csv("./Resources/rats.csv")
#cdataset = pd.read_csv("./Resources/Clean/clean_community_health.csv")
```

```python
# Expand output display to view all
pd.set_option('display.max_columns', 999)
```

```python
# Create DataFrame
rats_df = pd.DataFrame(dataset)
del rats_df['Unnamed: 0']
rats_df
```

```python
rats_df.loc[rats_df['Prop_Eng'] == '**']
```

```python
# Replace '**' with '0'
new_rats_df = rats_df.replace(to_replace = "**", value = 0)
new_rats_df.head()
```

```python
# Look up '**' in Prop-Eng
new_rats_df.loc[new_rats_df['Prop_Eng'] == '**']
```

```python
new_rats_df.loc[new_rats_df["Prop_obse"] == "**"]
```

```python
new_rats_df.drop(columns=["Unnamed: 0.1"])
```

```python
# Export to csv
new_rats_df.to_csv("./Resources/Clean/clean_new_rats.csv")
```
