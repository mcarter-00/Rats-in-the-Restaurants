### Data changes made in this notebook:
* Fixed the LAT and LNG columns (data values were flipped)
* Dropped 
    * Unnamed: 0
    * Geoname
    * PROGRAM_STATUS
    * violation_status
    * ACTIVITY_DATE
    * FACILITY_STATE
* Set program_status = active only
* Violations include:
    * F022
    * F023
    * F029
    * F038
    * F041
    * F051

```python
import pandas as pd
from datetime import datetime
```

```python
dataset = pd.read_csv("./Resources/Clean/clean_new_rats_JUNE09.csv")
```

```python
# Expand output display to view all
pd.set_option('display.max_columns', 999)
```

```python
df = pd.DataFrame(dataset)
df
```

```python
df.shape
```

```python
df1 = df.drop(columns = ["Unnamed: 0", "GEONAME", "ACTIVITY_DATE", "FACILITY_STATE"])
df1.head()
```

```python
df1.shape
```

```python
df1.rename(columns = {'LAT':'LNG', 
                      'LNG':'LAT'}, 
                       inplace = True)
df1.head()
```

```python
df2 = df1.loc[df['PROGRAM_STATUS'] == 'ACTIVE']
df2.head()
```

```python
df2.shape
```

```python
df3 = df2.loc[(df2['violation_code'] == 'F022') | (df2['violation_code'] == 'F023') |
             (df2['violation_code'] == 'F029') | (df2['violation_code'] == 'F038') | 
             (df2['violation_code'] == 'F041') | (df2['violation_code'] == 'F051')]
df3.head()
```

```python
df3.shape
```

```python
df3.dtypes
```

```python
df3.ACTIVITY_DATE_MONTHYEAR = pd.to_datetime(df3.ACTIVITY_DATE_MONTHYEAR)
```

```python
df3.dtypes
```

```python
df4 = df3.drop(columns = ["PROGRAM_STATUS", "violation_status"])
df4.head()
```

```python
df4.shape
```

```python
df4.to_csv("./Resources/Clean/clean_rats_JUNE12.csv")
```
